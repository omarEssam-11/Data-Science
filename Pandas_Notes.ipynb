```python
# getting most common developer type for each country and his percentage

#Note : The idxmax() function in Pandas is used to find the index of the first occurrence of the maximum value in a Series.



def get_most_common_developer(countries):
  data=[]
  for country in countries:
    most_common = (country_grp['DeveloperType'].value_counts(normalize=True).loc[country].idxmax())
    percent = (country_grp['DeveloperType'].value_counts(normalize=True).loc[country]).max() * 100
    data.append((country, most_common, percent))
  return pd.DataFrame(data, columns=['Country', 'Developer Type', 'Percentage'])



get_most_common_developer(['Egypt', 'Germany', 'India', 'United States'])
```

```python
def get_most_common_developer_all_countries(country_grp):
    data = []
    for country, group in country_grp:

        dev_type_counts = group['DeveloperType'].value_counts()

        if dev_type_counts.empty:
          continue

        most_common = dev_type_counts.idxmax()

        devs_no = dev_type_counts.max()
        data.append([country, most_common, devs_no])

    return pd.DataFrame(data, columns=['Country', 'Most Common Developer Type', 'No. of Developers']).sort_values(by='No. of Developers', ascending=False)



get_most_common_developer_all_countries(country_grp).head(14)

```
