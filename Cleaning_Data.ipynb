# **Data Cleaning**

# \#1 Dataset

```python
import pandas as pd
import numpy as np
import matplotlib as plt
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_excel('/content/drive/MyDrive/Datasets/Customer Call List.xlsx')
df
```

```python
#removing duplicates

df.drop_duplicates(inplace=True)
df
```

```python
# cleaning the last name

df['Last_Name'] = df['Last_Name'].str.strip('/_...')
df
```

```python
#standarize the phone number

#remove any special char
df["Phone_Number"] = df["Phone_Number"].str.replace('[^0-9]','', regex=True)

# remove NaN
df["Phone_Number"] = df["Phone_Number"].replace(np.NAN, '')

# make the number in this format 012-345-6789

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#extra cleaning

df["Phone_Number"] = df["Phone_Number"].str.replace('--','')

df
```

```python
# remove people who don't have numbers

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df= df.reset_index()
df
```

```python
#droping un useful col
df = df.drop('Not_Useful_Column', axis=1)
df
```

```python
# standrize yes => Y & No => N

df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('No', 'N')

df['Paying Customer'] = df['Paying Customer'].replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].replace('No', 'N')

df
```

```python
# drop people who do not want to be contacted

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
      df.drop(x, inplace=True)

df.drop(columns='index', inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
# splitting the address

df[['Street', 'State']] = df['Address'].str.split(',', expand=True)

df
```

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

df1 = pd.DataFrame(people)
df1
```

```python
#convert missing, Na , etc  ==> np.NAN
#so pd treats them as nan

df1.replace('Missing', np.NAN, inplace=True)

df1.replace('NA', np.NAN, inplace=True)

df1
```

```python
df1.drop_duplicates(inplace=True)
df1
```

```python
#drop row if it's all values are Nan
df1.dropna(axis='index', how='all', inplace=True)
df1
```

```python
#drop row if it's does not have first and last name
df1.dropna(axis='index', how='all', subset=['first', 'last'])
```

```python
#Casting

# df1['age'].astype(int)  =>> not gonna waork because of nan if a float not int

df1['age'] = df1['age'].astype(float)
df1.dtypes
```
